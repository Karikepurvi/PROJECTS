```python
df = pd.read_csv("Airbnb_Data.csv", engine='python')

df.head()
```

```python
df.info()
df.shape
df.describe()
```

```python

print(df.columns.tolist())


cols_to_drop = ['id', 'name', 'host_id', 'host_name', 'last_review','neighbourhood', 'latitude', 'longitude', 'thumbnail_url']

cols_to_drop = [col for col in cols_to_drop if col in df.columns]
df.drop(columns=cols_to_drop, inplace=True)
df.head()
```

```python
if 'review_scores_rating' in df.columns:
    df.loc[:, 'review_scores_rating'] = df['review_scores_rating'].fillna(df['review_scores_rating'].mean())

# Drop rows where log_price is null
df = df.loc[df['log_price'].notnull()]
df.reset_index(drop=True, inplace=True)
replace = {'log_price' : 'price'}
df.rename(columns=replace,inplace=True)
df.head()

```

```python
if 'accommodates' in df.columns:
    df['price_per_person'] = df['price'] / df['accommodates']

# Description length
if 'description' in df.columns:
    df['desc_length'] = df['description'].astype(str).apply(lambda x: len(x.split()))

# One-hot encoding for room_type and neighbourhood_group
categorical_cols = [col for col in ['room_type', 'neighbourhood_group'] if col in df.columns]
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

print("Final DataFrame Ready. Shape:", df.shape)
df.head()
```

```python
# 1. Import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# 2. Remove features derived from price
df_model = df.copy()

for col in ['price_per_person', 'log_price', 'predicted_price']:
    if col in df_model.columns:
        df_model.drop(columns=col, inplace=True)

# 3. Define target and features
target = 'price'
X = df_model.drop(columns=[target])
X = X.select_dtypes(include=[np.number])  # Only numeric features
y = df_model[target]

# Optional: Print selected features
print(" Features used:\n", X.columns.tolist())

# 4. Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=70
)

# 5. Train a smaller, more generalizable model
model = RandomForestRegressor(
    n_estimators=100, max_depth=8, random_state=70
)
model.fit(X_train, y_train)

# 6. Predict and evaluate
y_pred = model.predict(X_test)

rmse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)


print(f" RMSE: {rmse:.2f}")
print(f" RÂ² Score: {r2:.2f}")
```
